```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt



plt.figure(figsize=[12,6])
sns.set_style("whitegrid")

sns.lineplot(x, gureKDD,marker='D',markersize=10,color='red')
sns.lineplot(x, kddcup_neptune ,marker='P',markersize=11,color='limegreen')
sns.lineplot(x, kddcup ,marker='o',markersize=10,color='mediumorchid')
sns.lineplot(x, nsl_kdd ,marker='^',markersize=11,color='lightseagreen')
sns.lineplot(x, nb15 ,marker='d',markersize=12,color='darkslategray')

plt.xticks(x,fontsize=18)
plt.yticks(fontsize=18)
plt.ylim(0.25,1.05)
sns.despine()
patch_1 = mpatches.Patch(color='red', label='Gure-KDD')
patch_2 = mpatches.Patch(color='limegreen', label='KDDCup99-N')
patch_3 = mpatches.Patch(color='mediumorchid', label='KDDCup99')
patch_4 = mpatches.Patch(color='lightseagreen', label='NSL-KDD')
patch_5 = mpatches.Patch(color='darkslategray', label='UNSW-NB15')
plt.ylabel('Average Precision', fontsize=22)
plt.xlabel('Number of Negative Samples', fontsize=22)
plt.legend(handles=[patch_3,patch_2,patch_5,patch_4,patch_1], fontsize=20,loc='best')
plt.tight_layout()
plt.savefig('plot_neg_samples.pdf')
plt.show()
plt.close()
```
